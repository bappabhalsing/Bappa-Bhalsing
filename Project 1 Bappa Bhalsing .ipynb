# 1. Do Charges of people who smoke differ significantly from the people who don't?

# 2.Does BMI of males differ significantly from that of Females

```python
Male_BMI = df[df['sex'] == 'male'].bmi  #  BMI Male
Female_BMI = df[df['sex'] == 'female'].bmi #  BMI of Female

v1, v2 = np.var(Male_BMI) , np.var(Female_BMI)
print(v1,v2)
```

```python
n=1338 # size of population
dof=n-1 # degree of freedom
chi_critical= 46.19  # taken from data set.
```

```python
chi = (dof*v1)/v2

if chi < chi_critical:
    print("Since the test statistic is less than the critical value, we fail to reject the null")
else:
    print("Since the test statistic is more than the critical value, we reject the null")
```

# 3. Is the proportion of smokers significantly different in different genders?

```python
fsmokers = df[df['sex'] == 'female'].smoker.value_counts()[1]  # number of female smokers
msmokers = df[df['sex'] == 'male'].smoker.value_counts()[1] # number of male smokers
n_females = df.sex.value_counts()[1] # number of females in the data
n_males = df.sex.value_counts()[0] #number of males in the data
print([fsmokers, msmokers] , [n_females, n_males])
```

```python
print(f' Proportion of smokers in females, males = {round(115/662,2)}%, {round(159/676,2)}% respectively')
```

So Praportion of smoker is different in different genders.

```python
# Check signifance of smoker stastically
from statsmodels.stats.proportion import proportions_ztest
stat, pval = proportions_ztest([fsmokers, msmokers] , [n_females, n_males])
if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. \nWe reject the null Hypothesis')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. \nWe fail to reject the null Hypothesis')
```
